## Stemming

```python
words = ["builds", "building", "plays", "playing", "rides", "riding", "rider", "welding", "welded", "welder", "sitting", 'sat', "happiness"]
```

```python
## Using PorterStemmer() class
from nltk.stem import PorterStemmer
stemming = PorterStemmer()
```

```python
for word in words:
  print(word+"->"+stemming.stem(word))
```

```python
## Using RegexpStemmer() class
from nltk.stem import RegexpStemmer
reg_stemming = RegexpStemmer('s$|ing$', min=4)
```

```python
for word in words:
  print(word+"-->"+reg_stemming.stem(word))
```

```python
## Using SnowballStemmer() class
from nltk.stem import SnowballStemmer
snow_stemming = SnowballStemmer("english")
```

```python
for word in words:
  print(word+"-->"+snow_stemming.stem(word))
```

#### porter stem vs snow ball stem
* Showing how snowball stem performs better than porter stem

```python
print("For Porter Stemmer:", stemming.stem("sportingly"))
print("For Snowball Stemmer:", snow_stemming.stem("sportingly"))
```

#### Applying stopwords, followed by stemming

```python
para = """The Constitution of India is the supreme law that governs the country and defines the structure, powers, and duties of the government and the rights of citizens. It was adopted on November 26, 1949, and came into effect on January 26, 1950, which is celebrated as Republic Day. The Constitution was drafted by the Constituent Assembly of India, with B. R. Ambedkar serving as the chairman of the Drafting Committee. It establishes India as a sovereign, socialist, secular, democratic republic and guarantees fundamental rights such as equality, freedom, and justice to all citizens. The Constitution also outlines the responsibilities of the central and state governments and ensures the rule of law across the nation. It is one of the longest written constitutions in the world and plays a crucial role in maintaining democracy and unity in India."""
```

```python
sentences = nltk.sent_tokenize(para)
```

```python
for i in range(len(sentences)):
  words = nltk.word_tokenize(sentences[i])
  words = [stemmer.stem(word) for word in words if word not in set(stopwords.words('english'))]
  sentences[i] = ' '.join(words)
```

```python
sentences
```

### Using Lemmatizer

```python
para = """The Constitution of India is the supreme law that governs the country and defines the structure, powers, and duties of the government and the rights of citizens. It was adopted on November 26, 1949, and came into effect on January 26, 1950, which is celebrated as Republic Day. The Constitution was drafted by the Constituent Assembly of India, with B. R. Ambedkar serving as the chairman of the Drafting Committee. It establishes India as a sovereign, socialist, secular, democratic republic and guarantees fundamental rights such as equality, freedom, and justice to all citizens. The Constitution also outlines the responsibilities of the central and state governments and ensures the rule of law across the nation. It is one of the longest written constitutions in the world and plays a crucial role in maintaining democracy and unity in India."""
```

### Feature Engineering
* Categorizing the ratings as 0 (if rating less than 3) and 1 (if rating greater than 3)

```python
kindle_df.loc[:, "rating"] = kindle_df["rating"].apply(lambda a:0 if a<3 else 1)
```

### preprocessing the ratingText data

```python
kindle_df.loc[:, "reviewText"] = kindle_df["reviewText"].str.lower()
kindle_df["reviewText"]
```

* Cleaning the data by removing special characters

```python
import nltk
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
nltk.download('stopwords')
from bs4 import BeautifulSoup
```

```python
kindle_df.loc[:, 'reviewText']=kindle_df['reviewText'].apply(lambda x:re.sub('[^a-z A-z 0-9-]+', '',x))
## Removing the stopswords
kindle_df.loc[:, 'reviewText']=kindle_df['reviewText'].apply(lambda x:" ".join([y for y in x.split() if y not in stopwords.words('english')]))
## Removing url
kindle_df.loc[:, 'reviewText']=kindle_df['reviewText'].apply(lambda x: re.sub(r'(http|https|ftp|ssh)://([\w_-]+(?:(?:\.[\w_-]+)+))([\w.,@?^=%&:/~+#-]*[\w@?^=%&/~+#-])?', '' , str(x)))
## Removing html tags
kindle_df.loc[:, 'reviewText']=kindle_df['reviewText'].apply(lambda x: BeautifulSoup(x, 'lxml').get_text())
## Removing additional spaces
kindle_df.loc[:, 'reviewText']=kindle_df['reviewText'].apply(lambda x: " ".join(x.split()))
```

```python
kindle_df.head(10)
```

### Leammatization

```python
### Initializing
lemmatizer = WordNetLemmatizer()
```

```python
def word_lemmatizer(text):
  return " ".join([lemmatizer.lemmatize(word) for word in text.split()])
```

```python
kindle_df.loc[:, "reviewText"] = kindle_df["reviewText"].apply(lambda a:word_lemmatizer(a))
```

```python
kindle_df.head(10)
```

### Model building
* Splitting data to train and test sets

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(kindle_df["reviewText"], kindle_df["rating"], test_size=0.2)
```

```python
# BagofWords
from sklearn.feature_extraction.text import CountVectorizer
cv_bow = CountVectorizer()
X_train_bow = cv_bow.fit_transform(X_train).toarray()
X_test_bow = cv_bow.transform(X_test).toarray()
```

```python
# TF-IDF
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()
X_train_tfidf = tfidf.fit_transform(X_train).toarray()
X_test_tfidf = tfidf.transform(X_test).toarray()
```

```python
X_train_bow
```

```python
X_train_tfidf
```

* Using naive bayes algorithm as it performs well on sparse matrix (with almost zeroes)

```python
from sklearn.naive_bayes import GaussianNB
model_bow = GaussianNB().fit(X_train_bow, y_train)
model_tfidf = GaussianNB().fit(X_train_tfidf, y_train)
```

###Utilizng metrics

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
y_pred_bow = model_bow.predict(X_test_bow)
y_pred_tfidf = model_tfidf.predict(X_test_tfidf)
```

```python
print("Accuracy Score BOW: ", accuracy_score(y_test, y_pred_bow))
print("Accuracy Score TFIDF: ", accuracy_score(y_test, y_pred_tfidf))
```

```python
print("Confusion Matrix BOW: \n", confusion_matrix(y_test, y_pred_bow))
print("Confusion Matrix TFIDF: \n", confusion_matrix(y_test, y_pred_tfidf))
```

```python
from gensim.models import Word2Vec
```

```python
sentences = [text.split() for text in kindle_df["reviewText"]]   # tokenized reviews
```

```python
w2v = Word2Vec(sentences, vector_size=100, window=5, min_count=2)
```

```python
def w2v_avg(text):
    words = text.split()
    vectors = [w2v.wv[word] for word in words if word in w2v.wv]
    return np.mean(vectors, axis=0) if vectors else np.zeros(100)
```

```python
X_w2v = np.array([w2v_avg(text) for text in kindle_df['reviewText']])
```

```python
## Training
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

X_train, X_test, y_train, y_test = train_test_split(X_w2v, kindle_df["rating"], test_size=0.2)

model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
## prediction
y_pred_w2v = model.predict(X_test)
```

```python
print("Accuracy Score W2V: ", accuracy_score(y_test, y_pred_w2v))
```

```python
print("Confusion Matrix W2V: \n", confusion_matrix(y_test, y_pred_w2v))
```

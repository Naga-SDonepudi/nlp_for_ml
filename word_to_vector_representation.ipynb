### Bag of Words (BoW)
* A technique is NLP used to convert words to vector rpresenation based on the frequency of words
* Before changing words to vectors, it counts how many times each word appears in a sentence.

```python
import pandas as pd
```

```python
matter = pd.read_csv('SMSSpamCollection.txt',
                    sep='\t',names=["label","text"])
```

```python
matter
```

### Data Preprocessing

```python
import nltk
import re
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
```

```python
port_stemmer = PorterStemmer()
```

```python
corpus = []
for i in range(0, len(matter)):
  review=re.sub('[^a-zA-Z]',' ', matter['text'][i])
  review=review.lower()
  review=review.split()
  review=[port_stemmer.stem(word) for word in review if not word in stopwords.words('english')]
  review=' '.join(review)
  corpus.append(review)
```

### Word2Vec

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
```

```python
corpus = []
for i in range(0, len(matter)):
  review = re.sub('[^a-zA-Z]', ' ', matter['text'][i])
  review=review.lower()
  review = review.split()
  review=[lemmatizer.lemmatize(word) for word in review]
  review=' '.join(review)
  corpus.append(review)
```

```python
corpus
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
nltk.download('punkt_tab')
```

```python
words=[]
for sent in corpus:
  sent_token = sent_tokenize(sent)
  for sent in sent_token:
    words.append(simple_preprocess(sent))
```

```python
def avg_word2vec(doc):
  vectors = [model.wv[word] for word in doc if word in model.wv]

  if len(vectors) == 0:
    return np.zeros(model.vector_size)

  return np.mean(vectors, axis=0)
```
